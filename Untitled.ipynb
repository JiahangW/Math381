```python
import random
import numpy as np
import matplotlib.pyplot as plt

class Player:


    def __init__(self, score):
        self.score = score

    
def roll():
    return random.randint(1, 6) + random.randint(1, 6) # we use two six-sided dice here


numberOfPlayers = 2  ## the number of players
numberOfGames = 1000
goal = 50  # set the winning goal to be 50
#for i in range(numberOfPlayers):
#    wins.append(0)

```

```python
# codes for strategy A vs B
numberOfGames = 10000
AvsB = []
for j in range(-5, 5):
    scoreDiff = j # the score difference that we will change during the simulation
    singleComparison = []
    for times in range(10):
        print("This is ", times, " of ", j)
        wins = [] # keep track of the game result; reset after changing scoreDiff
        for i in range(numberOfPlayers):
            wins.append(0)
        for game in range(numberOfGames):
            # reset players in every game
            players = []
            for i in range(numberOfPlayers):
                players.append(0)

            # initialize each player with score of 0
            for i in range(numberOfPlayers):
                players[i] = Player(0)

            currentPlayer = game % numberOfPlayers # start with a different player each time
            higherScore = 0 # help track the status of the game
            while (higherScore < goal):
                if currentPlayer == 0:  # if this is the player using strategy A
                    # always add the rolls to the score
                    players[currentPlayer].score += roll()
                else:   # if this is the player using strategy B
                    # strategy B asks the player to keep track of the score difference
                    # if the opponent score is greater than player's score by 'scoreDiff'
                    # and the opponent score is larger than this roll (make sure no negative score appears)
                    # the player would subtract score instead of adding the score
                    # however, player would always add score if the player can win after doing so
                    result = roll()
                    if players[currentPlayer].score + result >= goal:
                        players[currentPlayer].score += result
                    elif (players[currentPlayer].score + scoreDiff < players[currentPlayer - 1].score and 
                        players[currentPlayer - 1].score >= result):
                        players[currentPlayer - 1].score -= result
                    else:
                        players[currentPlayer].score += result
                # update the higherScore after each roll
                higherScore = players[currentPlayer].score
                if players[currentPlayer-1].score > higherScore:
                        higherScore = players[currentPlayer-1].score
                # switch to the next player
                currentPlayer = (currentPlayer + 1) % numberOfPlayers

            # record the winner of this game
            for i in range(numberOfPlayers):
                if players[i].score == higherScore:
                    wins[i] += 1
        # add the winning rate of player 2 to the list
        singleComparison.append(wins[1]*1./numberOfGames)
    AvsB.append(singleComparison)
```

```python
# print out results for the previous part
for i in range(len(AvsB)):
    minAB = min(AvsB[i])
    maxAB = max(AvsB[i])
    print("The confidence interval of X = ", i-5, " is [", minAB, ", ", maxAB, "]")
```
